```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/asus/Downloads/Bank_dataset/cleaned_transactions.csv')
```

```python
df.head()
```

```python
# Monthly trend of total withdrawals
```

```python
df.groupby('YearMonth')['Total amount Withdrawn'].sum().plot(kind='line',figsize=(12,6),title = 'Monthly Total Withdrawals')
```

```python
# Top 5 ATMs by total withdrawals

top5 = df.groupby('ATM Name')['Total amount Withdrawn'].sum().nlargest(5)
top5.plot(kind='bar', figsize=(10,6),title='Top 5 ATMs')
```

```python
# Weekend vs Weekday withdrawals

df['date_type'] = df['Transaction Date'].dt.weekday.map(lambda x: 'Weekend' if x >=5 else 'Weekday')
sns.barplot(x='date_type', y ='Total amount Withdrawn', data = df, estimator = sum)
plt.title('Withdrawals: Weekend vs Weekday')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Weekend vs Weekday withdrawals
df['date_type'] = df['Transaction Date'].dt.weekday.map(
    lambda x: 'Weekend' if x > 5 else 'Weekday'
)

sns.barplot(
    x='date_type',
    y='Total amount Withdrawn',
    data=df,
    estimator= sum
)

plt.title('Withdrawals: Weekend vs Weekday')
plt.xlabel('Day Type')
plt.ylabel('Total Amount Withdrawn')
plt.tight_layout()
plt.show()
```

```python
df['Transaction Date'] = pd.to_datetime(
df['Transaction Date'], format= '%d/%m/%Y', errors = 'coerce'
)
```

```python
# XYZ vs Other card net difference over months

df.groupby('YearMonth')['net_xyz_vs_other_card'].sum().plot(kind ='bar', figsize(14,6), title = 'net_xyz_vs_other_card')
```
